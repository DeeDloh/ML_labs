8. Выведите минимальное и максимальное представимое значение для каждого скалярного типа numpy

````python
# np.int128, .... нет
# np.uint128 .... неи
types_int = [
    np.int8, np.int16, np.int32, np.int64,
    np.uint8, np.uint16, np.uint32, np.uint64
            ]
types_float = [np.float16, np.float32, np.float64]


print("Целочисленные типы")
for dtype in types_int:
    info = np.iinfo(dtype)
    print(dtype.__name__)
    print(f"min: {info.min}")
    print(f"max: {info.max}")
    print("-" * 32)
print()

print("Вещественные типы")
for dtype in types_float:
    info = np.finfo(dtype)
    print(dtype.__name__)
    print(f"min: {info.min}")
    print(f"max: {info.max}")
    print("-" * 32)
print()
# print(np.float16.max)
````

8. Правда ли, что у людей с индексом массы тела больше 75го перцентиля встречаются хронические заболевания чаще, чем у людей с индесом массы тела в переделах от 30 до 60 перцентиля?

Да
